```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Final.csv')
```

```python
df['Semestre'] = df['Semestre'].apply(lambda x: int(x.replace('º','').replace('°','')))
```

```python
df['Lun'] = df['Lun'].apply(lambda x: x.replace('\n','') if type(x) != float else x)
df['Mar'] = df['Mar'].apply(lambda x: x.replace('\n','') if type(x) != float else x)
df['Mie'] = df['Mie'].apply(lambda x: x.replace('\n','') if type(x) != float else x)
df['Jue'] = df['Jue'].apply(lambda x: x.replace('\n','') if type(x) != float else x)
df['Vie'] = df['Vie'].apply(lambda x: x.replace('\n','') if type(x) != float else x)
df['Link'] = df['Link'].apply(lambda x: np.nan if x == 'E' else x)
df['Calificacion'] = df['Calificacion'].apply(lambda x: np.nan if (x == 'R') or (x=='ERROR') else float(x))
```
